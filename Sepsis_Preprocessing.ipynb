```python
# Filter out columns have missing values greater than 95%
def preprocess_Data(input_dataframe):
     # Iterate over columns and check for the missing value percentage if 
    missing_percentage = (input_dataframe.isnull().sum() / input_dataframe.shape[0]) * 100
    print(missing_percentage)
    filtered_columns = list(missing_percentage[missing_percentage < 95].index)
    preprocessing_df = input_dataframe[filtered_columns]
    return preprocessing_df
```

```python
# Remaining columns in the database
preprocess_Train = preprocess_Data(train_data)
```
